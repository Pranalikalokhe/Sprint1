```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Load Data
df = pd.read_csv(r'E:\Internship2\ Ecommerse Bookstore\Final Year Project Ecommerse Bookstore\data\book_data.csv')

# Check data
print("Data shape:", df.shape)
print(df.head())

# Create binary target
df['ordered'] = df['orders'].apply(lambda x: 1 if x > 0 else 0)

# Features and target
X = df[['views']]
y = df['ordered']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,train_size = 0.8, random_state=42)

# Model
model = LogisticRegression()
model.fit(X_train, y_train)

# Evaluate
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))

# Predict probability of order for 5 views
print("Probability of order if views = 5:", model.predict_proba([[5]]))
```
